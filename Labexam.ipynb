```python
import random

GRID_SIZE = 3
VACUUM_CHAR = 'V'
DIRT_CHAR = 'D'
CLEAN_CHAR = 'C'

def print_grid(grid, vac_pos):
    print("-" * (GRID_SIZE * 4 + 1))
    for r_idx, row in enumerate(grid):
        row_str = "|"
        for c_idx, cell in enumerate(row):
            if (r_idx, c_idx) == vac_pos:
                row_str += f" {VACUUM_CHAR} |"
            elif cell:
                row_str += f" {DIRT_CHAR} |"
            else:
                row_str += f" {CLEAN_CHAR} |"
        print(row_str);
        print("-" * (GRID_SIZE * 4 + 1))

def count_clean_rows(grid):
    clean_rows_count = 0
    for r in range(GRID_SIZE):
        if all(not cell for cell in grid[r]):
            clean_rows_count += 1
    return clean_rows_count

def clean_3x3_square():
    grid = [[True for _ in range(GRID_SIZE)] for _ in range(GRID_SIZE)]
    vac_row, vac_col = random.randint(0, GRID_SIZE - 1), random.randint(0, GRID_SIZE - 1)
    vac_pos = (vac_row, vac_col)

    print("Initial State:")
    print_grid(grid, vac_pos)
    print(f"Vacuum starts at: {vac_pos}")
    print(f"Initial clean rows: {count_clean_rows(grid)}")
    print("\n-- Starting Cleaning Process --")

    steps = 0
    MAX_STEPS = GRID_SIZE * GRID_SIZE * 5

    while steps < MAX_STEPS:
        steps += 1
        vac_row, vac_col = vac_pos # Update vac_row, vac_col to current vac_pos
        print(f"\nStep {steps}: Vacuum at {vac_pos}")

        if grid[vac_row][vac_col]:
            simulated_grid = [list(row) for row in grid]
            simulated_grid[vac_row][vac_col] = False

            simulated_clean_rows_count = count_clean_rows(simulated_grid)

            if simulated_clean_rows_count > 2:
                print("CONSTRAINT ALERT: Cleaning the current cell would result in more than 2 fully clean rows.")
                print(f"Vacuum at current position: {vac_pos}. This action would make {simulated_clean_rows_count} rows clean.")
                print("Stopping cleaning to ensure the 'utmost 2 full rows clean' condition is met.")
                print("\n--- Final State (before violating action) ---")
                print_grid(grid, vac_pos)
                print(f"Total clean rows before violating action: {count_clean_rows(grid)}")
                break
            else:
                grid[vac_row][vac_col] = False
                print(f"Vacuum cleaned cell {vac_pos}.")
        else:
            print(f"Vacuum at {vac_pos} (already clean).")

        print_grid(grid, vac_pos)
        current_clean_rows_after_action = count_clean_rows(grid)
        print(f"Clean rows after action: {current_clean_rows_after_action}")

        all_clean = True;
        dirty_cells = []
        for r in range(GRID_SIZE):
            for c in range(GRID_SIZE):
                if grid[r][c]:
                    all_clean = False
                    dirty_cells.append((r, c))

        if all_clean:
            print("\nAll dirty cells have been cleaned. Stopping.")
            break

        if dirty_cells:
            min_dist = float('inf')
            next_target_row, next_target_col = vac_row, vac_col

            for dr, dc in dirty_cells:
                dist = abs(vac_row - dr) + abs(vac_col - dc)
                if dist < min_dist:
                    min_dist = dist
                    next_target_row, next_target_col = dr, dc

            new_vac_row, new_vac_col = vac_row, vac_col
            if next_target_row > vac_row:
                new_vac_row += 1
            elif next_target_row < vac_row:
                new_vac_row -= 1
            elif next_target_col > vac_col:
                new_vac_col += 1
            elif next_target_col < vac_col:
                new_vac_col -= 1

            vac_pos = (new_vac_row, new_vac_col)
            print(f"Vacuum moving to {vac_pos} (towards nearest dirty cell {next_target_row, next_target_col}).")
        else:
            print("No more dirty cells to move to. Stopping.")
            break

    if steps == MAX_STEPS:
        print(f"\nMax steps ({MAX_STEPS}) reached. Stopping.")

    print("\n-- Simulation Complete --")
    print("Final State of the Grid:")
    print_grid(grid, vac_pos)
    print(f"Total fully clean rows at end: {count_clean_rows(grid)}")
clean_3x3_square()
```
